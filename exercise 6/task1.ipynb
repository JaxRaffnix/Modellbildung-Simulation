```python
from OMPython import ModelicaSystem
import matplotlib.pyplot as plt

# !!! Cave: Make sure to run the python code in the same folder it is stored in.
# !!! Cave: Model name is case sensitive !
# !!! ompython dependency error distutils: solve by pip install setuptools
```

```python
modelname = "Circuit"
model = ModelicaSystem(modelname+".mo", modelname)

model.simulate()
```

```python
results = model.getSolutions()
[time] = model.getSolutions('time')
[coil_current] = model.getSolutions("inductor.i")
[coil_voltage] = model.getSolutions('inductor.v')
[source_current] = model.getSolutions('resistor.i')
```

```python
plt.plot(time, coil_current)
plt.title("Circuit Currents")
plt.legend()
plt.show()

plt.plot(time, source_current)
plt.title("Circuit Currents")
plt.legend()
plt.show()

plt.plot(time, coil_voltage)
plt.title("Circuit Currents")
plt.legend()
plt.show()
```
