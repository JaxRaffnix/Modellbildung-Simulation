```python
from OMPython import ModelicaSystem
import matplotlib.pyplot as plt
import numpy as np

# !!! Cave: Make sure to run the python code in the same folder it is stored in.
# !!! Cave: Model name is case sensitive !
# !!! ompython dependency error distutils: solve by pip install setuptools
```

```python
modelname = "fourquadrant"
model = ModelicaSystem(modelname+'.mo', modelname, ['./ModSimBib/package.mo'])

model.simulate()
```

```python
model.setSimulationOptions('stopTime=0.04')
model.simulate()
[time]=model.getSolutions('time')
[coil_current]=model.getSolutions('inductor1.i')
[voltage]=model.getSolutions('vierqst1.u_out')
```

```python
plt.plot(time, 10*coil_current, 'r')
plt.plot(time, voltage, 'darkgreen')
```
